# Gradient Descent

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Create a data set
```

```python
n = 100
np.random.seed(42)
X = 2*np.random.rand(n, 1) # U(0,1) obs
y = 4 + 3 * X + np.random.randn(n, 1)
```

```python
plt.scatter(X, y,color="k",s=5)
plt.xlabel("x", fontsize=13)
plt.ylabel("y",rotation=0,fontsize=13)
plt.grid()
```

### Linear regression using the formula $b = (X^{'}X)^{-1}(X^{'}y)$

```python
# add column of ones
X_b = np.c_[np.ones(n), X]
X_b[:5]
```

```python
b1 = np.dot(X_b.T,X_b)
b1 = np.linalg.inv(b1)
b2 = np.dot(X_b.T,y)
beta = np.dot(b1,b2)
beta
```

```python
yhat = np.dot(X_b,beta)
error = y - yhat
```

```python
# loss is the sum of squared errors
cost = np.dot(error.T,error)
cost
```

```python
plt.scatter(X, y,s=2,color='k')
plt.plot(X,yhat,'r',linewidth = 0.2)
plt.grid()
```

### Linear regression with sklearn

```python
model1 = LinearRegression().fit(X,y)
yhat = model1.predict(X)
error = y - yhat
np.dot(error.T,error)
```

```python
model1.intercept_, model1.coef_
```

### Linear regression using gradient descent

```python
eta = 0.001
```

```python
np.random.seed(1)
beta = np.random.randn(2,1)
print(beta[0],beta[1])
```

```python
for i in range(999):
    error = y - np.dot(X_b,beta)  
    gradients = np.dot(X_b.T,error)
    beta = beta + eta * gradients
    cost = np.dot(error.T,error)
```

```python
yhat = np.dot(X_b,beta)
```
